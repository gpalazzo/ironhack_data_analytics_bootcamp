```python
"""
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games

# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
"""

from random import choice
from math import ceil

#Setting variables
options = ['stone', 'paper', 'scissors']
max_games = 5
human_choice = ""
human_wins = 0
machine_wins = 0
ties = 0
plays = 0
stop = False

wins_to_end = ceil(max_games/2.0)
#another possibility would be divide by 2 and plus 0.5. it's gonna work since we know it must be an odd number
#wins_to_end = (max_games/2.0) + 0.5

wins_check = {'stone_paper': 1, #concating humanchoice_machinechoice
              'stone_scissors': 2,
              'paper_stone': 2,
              'paper_scissors': 1,
              'scissors_paper': 2,
              'scissors_stone': 1}


#Functions
def random_choice(options):
    return choice(options)

def check_winner(human_choice, machine_choice):
    return 0 if human_choice == machine_choice else wins_check[human_choice + "_" + machine_choice]


#Main Loop
while not plays > max_games and not stop:
    
    while not human_choice in options:
        print()
        human_choice = input("Choose your option: \n")
        machine_choice = random_choice(options)
    
    plays += 1
    print("Starting game: {}...".format(plays))
    print("Human's choice: {}\nMachine's choice: {}".format(human_choice, machine_choice))
    
    win_result = check_winner(human_choice, machine_choice)
    
    if win_result == 1:
        print("Machine wins.")
        machine_wins += 1
    elif win_result == 2:
        print("Human wins.")
        human_wins += 1
    else:
        print("Tie.")
        ties += 1
    
    human_choice = ""
        
    if machine_wins == wins_to_end or human_wins == wins_to_end:
        stop = True
    
print()
if machine_wins > human_wins:
    print("Machine is the winner!")

elif machine_wins < human_wins:
    print("Human is the winner!")

else:
    print("No winner, game tied.")
```
